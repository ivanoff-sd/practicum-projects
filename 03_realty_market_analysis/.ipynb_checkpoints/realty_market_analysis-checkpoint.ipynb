##### Пропуски в столбцах 'living_area', 'kitchen_area'

Данные столбцы следует рассматривать одновременно, принимая во внимание также показатели общей площади, количества комнат, studio и open_plan. Учитывая, что данные показатели находятся во взаимной зависимости, рекомендую удалить объекты, в которых не указан ни living_area, ни kitchen_area, при том не являющиеся студиями и квартирами свободной планировки. Пропуски kitchen_area для объектов свободной планировки и студий заполним значением "ноль", т.к. в исходных данных подавляющее большинство значений пропущено, соответственно кухни в данных объектах не предусмотрены.

```python
data = data.drop(data.query('living_area.isna() and kitchen_area.isna() and studio == False and open_plan == False').index)
#удалены объекты без living area и kitchen area, за исключением студий и открытых планировок.
data.loc[(data['studio']==True) & (data['kitchen_area'].isna()),'kitchen_area']=0
data.loc[(data['open_plan']==True) & (data['kitchen_area'].isna()),'kitchen_area']=0
# заполнены пропуски
print(data.loc[data['open_plan']==True]['kitchen_area'].isna().sum())
print(data.loc[data['studio']==True]['kitchen_area'].isna().sum())
# проверка заполнения пропусков
```

Для заполнения пропусков в living_area и kitchen_area на других видах объектов предлагаю следующий метод: рассчитать медианную долю, занимаемую жилыми комнатами и кухней на объектах всех классов (студии, однокомнатные, двух-, трёхкомнатные квартиры и т.д.) и помножить полученную долю на общую площадь объекта. Для этого рассчитаем показатели, сохраним их в отдельной таблице и воспользуемся функцией для одной строки.

```python
share_table = data.pivot_table(index = ['studio', 'open_plan', 'rooms'], values = ('total_area' , 'living_area', 'kitchen_area'), aggfunc = 'mean')
share_table['living_share'] = share_table['living_area'] / share_table['total_area']
share_table['kitchen_share'] = share_table['kitchen_area'] / share_table['total_area']
display(share_table)
```

Видим один пропуск в графе living_share. Заполним его предыдущим значением во избежание пропусков при заполнении данных. Изучим структуру индексов share_table, чтоб затем передать их в функцию, которая будет искать в share_table значения для заполнения пропусков.

```python
share_table['living_share'].fillna(value = 0.734, inplace=True)
display(share_table)
```

```python
share_table.index  # посмотрим, как устроены индексы в share_table, чтоб грамотно передать их в функцию по заполнению пропусков
```

```python
# объявим функцию, которая заполнит пропуски в значениях kitchen_area и living_area с учетом типа недвижимости

share_table.loc[(True, False, 0) , 'living_share']


def fill_areas(row):
    if pd.isna(row['living_area']):
        row['living_area'] = share_table.loc[
            (row['studio'], row['open_plan'], row['rooms']), 'living_share'] * row['total_area']
    if pd.isna(row['kitchen_area']):
        row['kitchen_area'] = share_table.loc[
            (row['studio'], row['open_plan'], row['rooms']), 'kitchen_share'] * row['total_area']
    return row

```

```python
data = data.apply(fill_areas, axis=1)
#применяем функцию для data
```

```python
data.isna().sum() # проверяем оставшиеся пропуски
```

Как видно выше, показатели living_area и kitchen_area полностью заполнены.

##### Пропуски в данных о парках

Пропуски в столбцах 'parks_nearest' и 'parks_around3000' будут обработаны следующим образом:
- из описания данных нам известно, что данные показатели рассчитаны автоматически на основании картографических данных.
- предположим, что эти данные не указаны в случае, если на расстоянии 3км от квартиры нет парков.
- убедимся, что при значении parks_around3000 == 0 или NaN parks_nearest пропущен либо превышает 3000.
- в случае подтверждения такой гипотезы заполним пропуски в parks_around3000 нулями, а в тех же строках заполним пропуски в parks_nearest условным значением 4000.
- пропуски в parks_nearest при наличии ненулевого значения parks_around3000 заполним медианными значениями, сгруппировав по столбцу parks_around3000.

```python
display(data[['parks_nearest','parks_around3000']].isna().sum())
# отображаем число пропущенных значений
data['parks_around3000'].value_counts()
# отображаем уникальные значения столбца
```

```python
display(data[['parks_nearest','parks_around3000']].query(
    'parks_around3000==0 or parks_around3000.isna()')['parks_nearest'].unique())
# отображаем значения расстояний до парка на объектах, где атрибут parks_around3000 не указан или равен 0, 
# чтобы убедиться: заполнение пропусков в 'parks_around3000' нулями не принесет искажений.
```

```python
data['parks_around3000'] = data['parks_around3000'].fillna(0) #заполнение пропусков нулями
print(data['parks_around3000'].value_counts()) # проверка заполнения
print(data['parks_around3000'].isna().sum())
data.loc[(data['parks_around3000']==0) & (data['parks_nearest'].isna()),'parks_nearest'] = 4000 # заполнение
 # пропусков в 'parks_nearest' условным значением 4000 для объектов, где 'parks_around3000' == 0.
print(data['parks_nearest'].isna().sum()) # проверка заполнения: число пропусков сократилось.
```

Для заполнения оставшихся пропусков медианными значениями по группам объявим функцию:

```python
def fill_median(data, groupby_column, filled_column):
    for value in data[groupby_column].unique():
        if not pd.isna(data.loc[(data[groupby_column]==value), filled_column].mean()):
            data.loc[(data[groupby_column]==value) & (data[filled_column].isna()),
                 filled_column] = data.loc[(data[groupby_column]==value), filled_column].median()

# данная функция призвана заполнить пропуски в filled_column медианным значением filling column,
# рассчитанным при группировке данных по значениям groupby_column
```

```python
fill_median(data, 'parks_around3000', 'parks_nearest')
data['parks_nearest'].hist()
# применение функции и вывод гистограммы - проверить, что изменилось
data.loc[:,['parks_around3000', 'parks_nearest']].isna().sum()
# подсчёт оставшихся пропусков
```

Если расстояние и наличие парков не указаны там, где поблизости их нет, то на 14615 объектах расстояние до ближайшего парка превышает 3000м.

```python
print(data[data['airports_nearest'].isna()]['locality_name'].value_counts().head(20))
print(data[data['cityCenters_nearest'].isna()]['locality_name'].value_counts().head(20))
```

Очевидно, большинство объектов, где пропущены эти значения, лежат за пределами Санкт-Петербурга. Это позволит нам заполнить пропуски медианным значением по населенному пункту там, где это доступно. Данное заполнение осуществимо лишь после устранения неявных дубликатов в названиях нас. пунктов, что и будет выполнено ниже.

#### Обработка дубликатов

```python
data.duplicated().sum()
```

Очевидно, явных дубликатов нет. Однако, нам необходимо устранить неявные дубликаты в названиях населенных пунктов, чтоб точнее заполнить пропущенные расстояния до центра города и аэродрома медианными значениями там, где эти значения будут представлены. Для начала еще раз посмотрим на список значений столбца 'locality_name'

```python
sorted(data['locality_name'].unique())
```

Из данного списка следует, что некоторые категории населенных пунктов записаны разными способами. С целью унификации формы названия и возможного объединения неявных дубликатов предложен способ с использованием списков значений, вложенных циклов и метода replace().

```python
# Объявляем функцию, которая будет исправлять типы населенных пунктов
def town_name_edit(str):
    for item in [
    'городской поселок',
    'городской посёлок',
    'поселок городского типа',
    'посёлок городского типа',
    'коттеджный поселок',
    'коттеджный посёлок',
    'посёлок при железнодорожной',
    'поселок',
    'посёлок'
            ]:
        if item in str:
            str = str.replace(item, 'пос.')
        
    for item in ['садоводческое некоммерческое товарищество', 'садовое товарищество']:
        if item in str:
            str = str.replace(item, 'СНТ')
    return str
```

```python
data['locality_name'] = data['locality_name'].apply(town_name_edit)
sorted(data['locality_name'].unique())
```

Встречается повтор названия: Мурино, пос. Мурино. Как известно, Мурино уже считается городом, так что исправим эту неточность.

```python
data.loc[data['locality_name']=='пос. Мурино', 'locality_name']='Мурино'
data.duplicated().sum()
```

Итак, неявные дубликаты в названиях нас. пунктов устранены, явные дубликаты отсутствуют.

#### Посчитайте среднюю цену одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений. Выделите населённые пункты с самой высокой и низкой стоимостью квадратного метра. 

```python
popular_towns = list(data.groupby('locality_name').count().sort_values(by='last_price', ascending=False).reset_index().loc[0:9, 'locality_name'])
data.loc[data['locality_name'].isin(popular_towns)].pivot_table(index='locality_name', values='m2_price', aggfunc='median').sort_values(by='m2_price')
```

Топ-10 населенных пунктов представлен выше. Наибольшая средняя цена за квадратный метр в Санкт-Петербурге, наименьшая - в Выборге.

#### выделите квартиры в Санкт-Петербурге с помощью столбца locality_name и вычислите среднюю цену каждого километра. Опишите, как стоимость объектов зависит от расстояния до центра города.

Для подсчёта "средней цены километра" построим сводную таблицу, где будет средняя стоимость квартир, сгруппированных по удалению от центра города.

```python
km_price_table = data.loc[data['locality_name']=='Санкт-Петербург'].pivot_table(index='cityCenters_km', values='last_price').reset_index()
display(km_price_table)
km_price_table.plot(kind='bar', x='cityCenters_km', y='last_price', figsize=(10,7), title = 'Зависимость цены недвижимости от расстояния до центра города')
```

На графике видно, что цена по мере удаления от центра города имеет тенденцию к снижению, но от 3 до 7 км она меняется неравномерно. Это можно связать с появлением новой элитной недвижимости поблизости от центра города. Тем не менее, для вычисления средней цены километра добавим в сводную таблицу столбец с разницей в средней цене между предыдущей и текущей строкой, а затем по данному столбцу вычислим среднее.

```python
for index in range(1, len(km_price_table)):
    
    km_price_table.loc[index, 'delta_price'] = km_price_table.loc[
        index-1, 'last_price'] - km_price_table.loc[index, 'last_price']
display(km_price_table)
print(km_price_table['delta_price'].mean())
```

Средняя цена километра составила около 969 тыс. рублей.
