```python
data.columns = data.columns.str.lower()
data.info()
```

Оценим количество явных дубликатов:

```python
data.duplicated().sum()
```

Полных дубликатов не обнаружено. В таком случае, поищем дубликаты в сочетаниях "название игры - платформа - год выхода", которое должно однозначно определять игру:

```python
display(data.duplicated(subset=['name','platform','year_of_release']).sum())
data = data.drop_duplicates(subset=['name','platform']).reset_index(drop=True)

display(data.duplicated(subset=['name','platform']).sum())
```

### Приведение типов данных

```python
data.info()
```

Ниже столбец 'year_of_release' будет приведен к необходимому типу позже, после того как будет осуществлена замена пропущенных значений. 

Несмотря на то, что столбец 'critic_score' содержит целые значения, привести его к integer не получится, т.к. в нем присутствует NaN - значение типа float.

Столбец 'user_score' может быть приведен к типу float, если удалить из него значения 'tbd' типа string. Учитывая, что tbd - To be Defined - означает, что рейтинг еще не определён, смело можно заменить это значение на NaN.

Прочие столбцы имеют тип, соответствующий их содержанию.


```python
# удаляем значения "tbd" из столбца 'user_score':
data.loc[data['user_score'] == 'tbd', 'user_score'] = np.nan
# приводим 'user_score' к типу float64:
data['user_score'] = data['user_score'].astype(float)

data.info()
```

#### Пропуски 'critic_score',  'user_score' и 'rating'

Обращает на себя внимание большое количество пропусков в столбцах:

critic_score

user_score

rating

Около половины игр не имеют данных показателей. Даже с учётом того, что число записей в user_score было уменьшено в процессе предобработки, нельзя объяснить эту особенность. Необходимо разобраться в природе данных пропусков.

Предположим, что это связано с годом выхода игры: возможно, в 80-е и 90-е не было принято оценивать игры таким образом, чтоб это могло попасть в наши данные. Для этого посчитаем количество записей, в которых все три показателя пропущены, и сгруппируем по году выхода, полученный Series разделим на общее количество записей, сгруппированных по году выхода, а результат представим в виде barplot-диаграммы.

```python
(data.loc[(data['critic_score'].isna()) & (data['user_score'].isna()) & (data['rating'].isna()), 'year_of_release'].value_counts()/data['year_of_release'].value_counts()).plot(kind='bar', figsize = (15, 7), xlabel = 'Год выхода', ylabel = "Доля игр без рейтингов доля игр без рейтингов по году выпуска");
```

Действительно, согласно диаграмме, вплоть до 1999 года абсолютное большинство игр не имело ни одного из трёх рейтингов. 
Однако, со временем данная проблема не исчезла: в последующие годы доля табличных записей без рейтинга составляет приблизительно от четверти до половины. 

Предположим, что это связано с платформой, для которой выпустили игру. Что если не для каждой платформы существуют эти рейтинги? А если игра выходит для нескольких платформ, рейтинг указывается один раз или в каждой строке?

Чтобы в этом разобраться, составим аналогичный график, сгруппировав строки таблицы по платформам, а затем ознакомимся с табличными сведениями для кросс-платформенных игр.

```python
(data.loc[
    (data['critic_score'].isna()) & (data['user_score'].isna() & (data['rating'].isna())), 'platform'
].value_counts() / data['platform'].value_counts()).sort_values(ascending=False)\
.plot(kind='bar', figsize = (15, 7), xlabel = 'Платформа', ylabel = "Доля игр без рейтингов", title='доля игр без рейтингов по игровым платформам');
```

Для 14 платформ доля игр без рейтингов составила 100%. Соберём года выхода игр для этих платформ:

```python
# Получим долю игр без рейтингов аналогично предыдущему выводу: 
platforms = data.loc[
    (data['critic_score'].isna()) & (data['user_score'].isna() & (data['rating'].isna())), 'platform'
].value_counts() / data['platform'].value_counts()
# Сохраним список платформ, где доля игр без рейтингов == 1 (то есть 100%)
platforms = platforms[platforms==1].index.to_list()
# Посмотрим, в какие годы выходили игры на этих платформах
sorted(data.loc[data['platform'].isin(platforms), 'year_of_release'].unique())
```

Действительно, игры на данных платформах выходили до 2002 года включительно, что соотносится с предыдущим выводом. Теперь посмотрим, что с кросс-платформенными играми. Сохраним список из 10 игр, название которых встречается чаще всего, что должно быть связано с количеством платформ, на которых они вышли. Затем выведем табличные данные по этим играм.

```python
games = data['name'].value_counts().head(10).index.to_list()
for item in games:
    display(data.loc[data['name']==item].head())
```

На основании полученных данных можно предположить, что рейтинги одной и той же игры на разных платформах не совпадают между собой и отсутствовать могут независимо друг от друга. В таком случае, попробуем визуализировать еще одно предположение: исключим игры до 2000 г., то есть той эпохи, по которой данных о рейтингах недостаточно, и после 2015 года, которые на момент получения данных были новыми и могли еще не иметь рейтинга. Для оставшихся игр суммируем их продажи по всем регионам и узнаем, характерны ли пропуски в рейтинге для игр с малыми объемами продаж.

```python
sample = data.loc[(data['year_of_release']< 2016 ) & (data['year_of_release']>1999)].reset_index(drop=True)
sample['total_sales'] = sample['na_sales'] + sample['eu_sales'] + sample['jp_sales'] + sample['other_sales']
sample['critic_isna'] = sample['critic_score'].isna()
sample['user_isna'] = sample['user_score'].isna()
sample['rating_isna'] = sample['rating'].isna()


sample = sample.pivot_table(index=['critic_isna','user_isna', 'rating_isna'], values='total_sales', aggfunc = 'median')
sample.plot(kind = 'bar', xlabel = "no critic_score, no user_score, no rating", figsize = [10, 5], title='Продажи игр, сгруппированные по наличию/отсутствию рейтингов');
```

На диаграмме видно, что игры, у которых указан user_score и rating, продавались значительно чаще остальных. Среди них игры, имеющие оценку критиков, продавались лучше, чем не имеющие таковой. Таким образом, сделаем вывод, что пропуски в столбцах 'critic_score', 'user_score' и 'rating' связаны с годом выхода и популярностью игры. Надежных данных, чтоб заполнить пропуски в оценке критиков и пользователей у нас нет. Также, заполнять пропуски в ESRB значением 'pending' было бы неосмотрительно - мы не знаем, происходит ли установка рейтинга в данное время, или же он вообще не рассматривался, или пропущен только в наших данных. Вывод - адекватно заполнить данные пропуски мы не можем. 

#### Пропуски значений 'name', 'genre', 'year_of_release'
Название и жанр имеют лишь по одному пропуску. Строки, содержащие данные значения, будут удалены. Пропуски года выхода игры можно заполнить с известным допущением, если игра кроссплатформенная, и мы знаем, когда вышли ее другие версии.

```python
display(data.loc[data.loc[:,'name'].isna()])
# пропуски в названии и жанре находятся в одной строке
data = data.dropna(subset=['name']).reset_index(drop=True)
# удалены пропуски в названии и жанре
```

```python
names = data.loc[data['year_of_release'].isna(), 'name'].unique()
names_years_table = data.loc[data['name'].isin(names)].groupby('name')['year_of_release'].agg('median')
print(names_years_table.isna().sum())
names_years_table.dropna(inplace=True)
len(names_years_table)
```

Для 98 игр "нашелся" год выпуска. Заполним пропущенные значения с помощью данных таблицы. Оставшиеся игры без года выпуска - удалим, после чего столбец с годом выпуска игры приведем к типу int.

```python
for name in names_years_table.index:
    data.loc[(data.loc[:, 'name']==name) & (data.loc[:, 'year_of_release'].isna()),'year_of_release']=\
    names_years_table[name]
```

```python
data = data.dropna(subset=['year_of_release']).reset_index(drop=True)
print(data.isna().sum())
data['year_of_release'] = data['year_of_release'].astype(int)
```

### Динамика распределения продаж по платформам
#### Исследование динамики
Для исследования динамики использования игровых платформ возьмем срез данных, начиная от 1990. 


```python
data = data.loc[data['year_of_release']>1989] # удалены данные до 1990

# создадим список из 20 платформ с наибольшим суммарным числом продаж

platforms = data.groupby('platform')['total_sales'].agg('sum').sort_values().tail(20).index.to_list()

# построим график, описывающий распределение игр на 20 популярных платформах по годам выпуска

data.loc[data.loc[:,'platform'].isin(platforms), ['platform', 'total_sales','year_of_release']].pivot_table(index=
'year_of_release', columns = 'platform', values = 'total_sales', aggfunc='sum').plot(figsize = (15,7), ylabel='total sales', title = "динамика использования игровых платформ во времени");
```

График иллюстрирует, что большинство популярных платформ за исключением ПК остаются актуальными на протяжение не более, чем 10 лет. Для визуализации актуальных на момент исследования тенденций сделаем срез данных от 2010 года до и оставим 10 самых популярных платформ. 

```python
platforms = data.loc[data['year_of_release']>2009].groupby('platform')['total_sales'].agg('sum').sort_values().tail(10).index.to_list()
# создаем список актуальных платформ с 2010

data.loc[(data.loc[:,'platform'].isin(platforms)) & (data.loc[:, 'year_of_release']>2009), ['platform', 'total_sales','year_of_release']].pivot_table(index=
'year_of_release', columns = 'platform', values = 'total_sales', aggfunc='sum').plot(figsize = (15,7), ylabel='total sales', title = "динамика использования игровых платформ во времени");
```

На полученном графике мы видим, что как минимум 3 платформы на 2015 год утратили свою актуальность, а данные на 2016 год показывают тотальное падение продаж. Из описания датасета нам известно, что эти данные неполные. Потому изменим график еще раз: возьмём 7 платформ, наиболее популярных c 2012 по 2015 и уберем из графика данные за неполный 2016 год.

```python
platforms = data.loc[data['year_of_release'].isin([2012, 2013,2014, 2015])].groupby('platform')['total_sales'].\
agg('sum').sort_values().tail(7).index.to_list()
# создаем список актуальных платформ с 2010

data.loc[(data.loc[:,'platform'].isin(platforms)) & (data.loc[:, 'year_of_release']>2009) & (data.loc[:, 'year_of_release']<2016), ['platform', 'total_sales','year_of_release']].pivot_table(index=
'year_of_release', columns = 'platform', values = 'total_sales', aggfunc='sum').plot(figsize = (15,7), ylabel='total sales', title = "динамика использования игровых платформ с 2010 по 2015");
```

### Влияние оценок пользователей и критиков на продажи игр
####  Расчёт для наиболее популярной платформы

```python
# Выявим наиболее успешную платформу за последние 5 лет:
best_platform = data.groupby('platform')['total_sales'].agg(sum).sort_values().index.to_list()[-1]

# Объявим функцию, которая будет группировать данные по платформам, считать корреляцию и строить scatterplotы
def score_sale_corr(df, platform):
    sample = df.loc[df['platform']==platform,['total_sales', 'critic_score', 'user_score']]
    print('Корреляция рейтинга и продаж для платформы', platform)
    print(sample.corr())
    sample.plot(kind='scatter', x='total_sales', y='critic_score', \
                xlim = (-0.5, 6), ylim=(30,95), alpha = 0.7, figsize = (15, 6))
    plt.show()
    sample.plot(kind='scatter', x='total_sales', y='user_score',\
                xlim = (-0.5, 6), ylim=(0.2,9.5), alpha = 0.7, figsize = (15, 6))
    plt.show()
score_sale_corr(data, best_platform)


```

На основании полученных данных можно судить о том, что оценка критиков имеет слабую положительную корреляцию с объёмом продаж и положительную корреляцию с оценкой игроков. Других закономерностей не выявлено. 
Необходимо рассмотреть аналогичные показатели для других платформ. Возьмем еще 5 в порядке убывания популярности.
#### Расчёт для прочих платформ

```python
platforms = data.groupby('platform')['total_sales'].agg(sum).sort_values().index.to_list()[-6: -1]
# получили список платформ

for platform in platforms:
    score_sale_corr(data, platform)
```

Во всех случаях можно отметить слабую положительную корреляцию между оценкой критиков и продажами игры и положительную корреляцию между оценкой пользователей и критиков. Других закономерностей не установлено.

## Анализ поведения пользователей по регионам

### Популярные платформы

```python
sample = data.pivot_table(index='platform', values =['na_sales', 'eu_sales', 'jp_sales', 'other_sales'],aggfunc='sum')
# создана таблица продаж, сгруппированных по платформе, по всем регионам.
sum_sample = sample.sum() 
# получены суммарные значения столбцов для перевода абсолютных значений в доли

for item in list(sample.columns):
    sample[item] = sample[item] / sum_sample[item] * 100 # переводим абсолютные значения в проценты
    
```

```python
display(sample)
sample.plot(kind='bar',figsize = (16, 6));
```

Как видно на графике, использование различных консолей сильно зависит от региона, что может быть связано как с объемом продаж консоли в разных странах, так и с приверженностью местных пользователей к определенному формату платформ. Ниже выясним подробнее, какие платформы популярны в представленных регионах и визуализируем доли покупок игр для этих платформ.  

```python
def get_top_5(df, column):
    series = df[column].sort_values()
    series['other']=100 - series.tail(5).sum()
    series=series.tail(6)
    print('Расчёт top-5 для', column)
    print(series)
    series.plot(kind='pie', ylabel='Доля продаж в процентах', title=column, grid=True, figsize=(10,10));
    plt.show()

regions = sample.columns.to_list()

for region in regions:
    get_top_5(sample, region)
```

На основании полученных распределений можно утверждать, что в Японии наиболее популярна консоль 3DS с большим отрывом от конкурентов, в Северной Америке - X360, за которым следуют PS3, PS4, XOne с небольшим отрывом, в других рассматриваемых регионах лидируют PS4, и за ним с небольшой разницей PS3.

## Статистический анализ
### Сравнение среднего пользовательского рейтинга платформ Xbox One и PC

Оценим размеры выборок. Если они значительно превышают 30 экземпляров и сопоставимы по количеству, воспользуемся критерием Стьюдента для 2х независимых выборок.

- H0: Средние пользовательские рейтинги для платформ Xbox One и PC одинаковы.
- Ha: Средние пользовательские рейтинги для платформ Xbox One и PC не одинаковы.



```python
group_1 = data.loc[data.loc[:, 'platform']=='XOne', 'user_score'].dropna().to_list()
print('размер выборки для XOne', len(group_1))
group_2 = data.loc[data.loc[:, 'platform']=='PC', 'user_score'].dropna().to_list()
print('размер выборки для PC', len(group_2))

results = st.ttest_ind(group_1, group_2)
print(results.pvalue)
if results.pvalue<0.05:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

### Сравнение среднего пользовательского рейтинга игр жанра Action и Sport

Оценим размеры выборок. Если они значительно превышают 30 экземпляров и сопоставимы по количеству, воспользуемся критерием Стьюдента для 2х независимых выборок. Учитывая, что размеры выборок различаются (531 и 194), воспользуемся параметром equal_var=False.

- H0: Средние пользовательские рейтинги игр жанра Action и Sport одинаковы.
- Ha: Средние пользовательские рейтинги игр жанра Action и Sport не одинаковы.

```python
group_1 = data.loc[data.loc[:, 'genre']=='Action', 'user_score'].dropna().to_list()
print('размер выборки для Action', len(group_1))
group_2 = data.loc[data.loc[:, 'genre']=='Sports', 'user_score'].dropna().to_list()
print('размер выборки для Sports', len(group_2))

results = st.ttest_ind(group_1, group_2, equal_var=False)
print(results.pvalue)
if results.pvalue<0.05:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
    
print('Средняя оценка игр в жанре Action', np.mean(group_1))
print('Средняя оценка игр в жанре Sports', np.mean(group_2))
```

### Результаты

- Средние пользовательские рейтинги игр жанра Action и Sport одинаковы.
- Средние пользовательские рейтинги игр жанра Action и Sport не одинаковы.
