–î–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ, —Å–æ–≥–ª–∞—Å–Ω–æ –¥–∏–∞–≥—Ä–∞–º–º–µ, –≤–ø–ª–æ—Ç—å –¥–æ 1999 –≥–æ–¥–∞ –∞–±—Å–æ–ª—é—Ç–Ω–æ–µ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –∏–≥—Ä –Ω–µ –∏–º–µ–ª–æ –Ω–∏ –æ–¥–Ω–æ–≥–æ –∏–∑ —Ç—Ä—ë—Ö —Ä–µ–π—Ç–∏–Ω–≥–æ–≤. 
–û–¥–Ω–∞–∫–æ, —Å–æ –≤—Ä–µ–º–µ–Ω–µ–º –¥–∞–Ω–Ω–∞—è –ø—Ä–æ–±–ª–µ–º–∞ –Ω–µ –∏—Å—á–µ–∑–ª–∞: –≤ –ø–æ—Å–ª–µ–¥—É—é—â–∏–µ –≥–æ–¥—ã –¥–æ–ª—è —Ç–∞–±–ª–∏—á–Ω—ã—Ö –∑–∞–ø–∏—Å–µ–π –±–µ–∑ —Ä–µ–π—Ç–∏–Ω–≥–∞ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ –æ—Ç —á–µ—Ç–≤–µ—Ä—Ç–∏ –¥–æ –ø–æ–ª–æ–≤–∏–Ω—ã. 

–ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, —á—Ç–æ —ç—Ç–æ —Å–≤—è–∑–∞–Ω–æ —Å –ø–ª–∞—Ç—Ñ–æ—Ä–º–æ–π, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–π –≤—ã–ø—É—Å—Ç–∏–ª–∏ –∏–≥—Ä—É. –ß—Ç–æ –µ—Å–ª–∏ –Ω–µ –¥–ª—è –∫–∞–∂–¥–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å—É—â–µ—Å—Ç–≤—É—é—Ç —ç—Ç–∏ —Ä–µ–π—Ç–∏–Ω–≥–∏? –ê –µ—Å–ª–∏ –∏–≥—Ä–∞ –≤—ã—Ö–æ–¥–∏—Ç –¥–ª—è –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º, —Ä–µ–π—Ç–∏–Ω–≥ —É–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è –æ–¥–∏–Ω —Ä–∞–∑ –∏–ª–∏ –≤ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ?

–ß—Ç–æ–±—ã –≤ —ç—Ç–æ–º —Ä–∞–∑–æ–±—Ä–∞—Ç—å—Å—è, —Å–æ—Å—Ç–∞–≤–∏–º –∞–Ω–∞–ª–æ–≥–∏—á–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫, —Å–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞–≤ —Å—Ç—Ä–æ–∫–∏ —Ç–∞–±–ª–∏—Ü—ã –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º, –∞ –∑–∞—Ç–µ–º –æ–∑–Ω–∞–∫–æ–º–∏–º—Å—è —Å —Ç–∞–±–ª–∏—á–Ω—ã–º–∏ —Å–≤–µ–¥–µ–Ω–∏—è–º–∏ –¥–ª—è –∫—Ä–æ—Å—Å-–ø–ª–∞—Ç—Ñ–æ—Ä–º–µ–Ω–Ω—ã—Ö –∏–≥—Ä.

–î–ª—è 14 –ø–ª–∞—Ç—Ñ–æ—Ä–º –¥–æ–ª—è –∏–≥—Ä –±–µ–∑ —Ä–µ–π—Ç–∏–Ω–≥–æ–≤ —Å–æ—Å—Ç–∞–≤–∏–ª–∞ 100%. –°–æ–±–µ—Ä—ë–º –≥–æ–¥–∞ –≤—ã—Ö–æ–¥–∞ –∏–≥—Ä –¥–ª—è —ç—Ç–∏—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º:

```python
# –ü–æ–ª—É—á–∏–º –¥–æ–ª—é –∏–≥—Ä –±–µ–∑ —Ä–µ–π—Ç–∏–Ω–≥–æ–≤ –∞–Ω–∞–ª–æ–≥–∏—á–Ω–æ –ø—Ä–µ–¥—ã–¥—É—â–µ–º—É –≤—ã–≤–æ–¥—É: 
platforms = data.loc[
    (data['critic_score'].isna()) & (data['user_score'].isna() & (data['rating'].isna())), 'platform'
].value_counts() / data['platform'].value_counts()
# –°–æ—Ö—Ä–∞–Ω–∏–º —Å–ø–∏—Å–æ–∫ –ø–ª–∞—Ç—Ñ–æ—Ä–º, –≥–¥–µ –¥–æ–ª—è –∏–≥—Ä –±–µ–∑ —Ä–µ–π—Ç–∏–Ω–≥–æ–≤ == 1 (—Ç–æ –µ—Å—Ç—å 100%)
platforms = platforms[platforms==1].index.to_list()
# –ü–æ—Å–º–æ—Ç—Ä–∏–º, –≤ –∫–∞–∫–∏–µ –≥–æ–¥—ã –≤—ã—Ö–æ–¥–∏–ª–∏ –∏–≥—Ä—ã –Ω–∞ —ç—Ç–∏—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö
sorted(data.loc[data['platform'].isin(platforms), 'year_of_release'].unique())
```

–î–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ, –∏–≥—Ä—ã –Ω–∞ –¥–∞–Ω–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö –≤—ã—Ö–æ–¥–∏–ª–∏ –¥–æ 2002 –≥–æ–¥–∞ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ, —á—Ç–æ —Å–æ–æ—Ç–Ω–æ—Å–∏—Ç—Å—è —Å –ø—Ä–µ–¥—ã–¥—É—â–∏–º –≤—ã–≤–æ–¥–æ–º. –¢–µ–ø–µ—Ä—å –ø–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ —Å –∫—Ä–æ—Å—Å-–ø–ª–∞—Ç—Ñ–æ—Ä–º–µ–Ω–Ω—ã–º–∏ –∏–≥—Ä–∞–º–∏. –°–æ—Ö—Ä–∞–Ω–∏–º —Å–ø–∏—Å–æ–∫ –∏–∑ 10 –∏–≥—Ä, –Ω–∞–∑–≤–∞–Ω–∏–µ –∫–æ—Ç–æ—Ä—ã—Ö –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è —á–∞—â–µ –≤—Å–µ–≥–æ, —á—Ç–æ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Å–≤—è–∑–∞–Ω–æ —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –ø–ª–∞—Ç—Ñ–æ—Ä–º, –Ω–∞ –∫–æ—Ç–æ—Ä—ã—Ö –æ–Ω–∏ –≤—ã—à–ª–∏. –ó–∞—Ç–µ–º –≤—ã–≤–µ–¥–µ–º —Ç–∞–±–ª–∏—á–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –ø–æ —ç—Ç–∏–º –∏–≥—Ä–∞–º.

```python
games = data['name'].value_counts().head(10).index.to_list()
for item in games:
    display(data.loc[data['name']==item].head())
```

–ù–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –º–æ–∂–Ω–æ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç—å, —á—Ç–æ —Ä–µ–π—Ç–∏–Ω–≥–∏ –æ–¥–Ω–æ–π –∏ —Ç–æ–π –∂–µ –∏–≥—Ä—ã –Ω–∞ —Ä–∞–∑–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö –Ω–µ —Å–æ–≤–ø–∞–¥–∞—é—Ç –º–µ–∂–¥—É —Å–æ–±–æ–π –∏ –æ—Ç—Å—É—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –º–æ–≥—É—Ç –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞. –í —Ç–∞–∫–æ–º —Å–ª—É—á–∞–µ, –ø–æ–ø—Ä–æ–±—É–µ–º –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –µ—â–µ –æ–¥–Ω–æ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–µ: –∏—Å–∫–ª—é—á–∏–º –∏–≥—Ä—ã –¥–æ 2000 –≥., —Ç–æ –µ—Å—Ç—å —Ç–æ–π —ç–ø–æ—Ö–∏, –ø–æ –∫–æ—Ç–æ—Ä–æ–π –¥–∞–Ω–Ω—ã—Ö –æ —Ä–µ–π—Ç–∏–Ω–≥–∞—Ö –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ, –∏ –ø–æ—Å–ª–µ 2015 –≥–æ–¥–∞, –∫–æ—Ç–æ—Ä—ã–µ –Ω–∞ –º–æ–º–µ–Ω—Ç –ø–æ–ª—É—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –±—ã–ª–∏ –Ω–æ–≤—ã–º–∏ –∏ –º–æ–≥–ª–∏ –µ—â–µ –Ω–µ –∏–º–µ—Ç—å —Ä–µ–π—Ç–∏–Ω–≥–∞. –î–ª—è –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –∏–≥—Ä —Å—É–º–º–∏—Ä—É–µ–º –∏—Ö –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –≤—Å–µ–º —Ä–µ–≥–∏–æ–Ω–∞–º –∏ —É–∑–Ω–∞–µ–º, —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω—ã –ª–∏ –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —Ä–µ–π—Ç–∏–Ω–≥–µ –¥–ª—è –∏–≥—Ä —Å –º–∞–ª—ã–º–∏ –æ–±—ä–µ–º–∞–º–∏ –ø—Ä–æ–¥–∞–∂.

```python
sample = data.loc[(data['year_of_release']< 2016 ) & (data['year_of_release']>1999)].reset_index(drop=True)
sample['total_sales'] = sample['na_sales'] + sample['eu_sales'] + sample['jp_sales'] + sample['other_sales']
sample['critic_isna'] = sample['critic_score'].isna()
sample['user_isna'] = sample['user_score'].isna()
sample['rating_isna'] = sample['rating'].isna()


sample = sample.pivot_table(index=['critic_isna','user_isna', 'rating_isna'], values='total_sales', aggfunc = 'median')
sample.plot(kind = 'bar', xlabel = "no critic_score, no user_score, no rating", figsize = [10, 5], title='–ü—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä, —Å–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –ø–æ –Ω–∞–ª–∏—á–∏—é/–æ—Ç—Å—É—Ç—Å—Ç–≤–∏—é —Ä–µ–π—Ç–∏–Ω–≥–æ–≤');
```

#### –ò—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –¥–∏–Ω–∞–º–∏–∫–∏
–î–ª—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –¥–∏–Ω–∞–º–∏–∫–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –∏–≥—Ä–æ–≤—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –≤–æ–∑—å–º–µ–º —Å—Ä–µ–∑ –¥–∞–Ω–Ω—ã—Ö, –Ω–∞—á–∏–Ω–∞—è –æ—Ç 1990.

```python
# –í—ã—è–≤–∏–º –Ω–∞–∏–±–æ–ª–µ–µ —É—Å–ø–µ—à–Ω—É—é –ø–ª–∞—Ç—Ñ–æ—Ä–º—É –∑–∞ –ø–æ—Å–ª–µ–¥–Ω–∏–µ 5 –ª–µ—Ç:
best_platform = data.groupby('platform')['total_sales'].agg(sum).sort_values().index.to_list()[-1]

# –û–±—ä—è–≤–∏–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º, —Å—á–∏—Ç–∞—Ç—å –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é –∏ —Å—Ç—Ä–æ–∏—Ç—å scatterplot—ã
def score_sale_corr(df, platform):
    sample = df.loc[df['platform']==platform,['total_sales', 'critic_score', 'user_score']]
    print('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Ä–µ–π—Ç–∏–Ω–≥–∞ –∏ –ø—Ä–æ–¥–∞–∂ –¥–ª—è –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã', platform)
    print(sample.corr())
    sample.plot(kind='scatter', x='total_sales', y='critic_score', \
                xlim = (-0.5, 6), ylim=(30,95), alpha = 0.7, figsize = (15, 6))
    plt.show()
    sample.plot(kind='scatter', x='total_sales', y='user_score',\
                xlim = (-0.5, 6), ylim=(0.2,9.5), alpha = 0.7, figsize = (15, 6))
    plt.show()
score_sale_corr(data, best_platform)


```

```python
sample = data.groupby('genre')['total_sales'].agg(['sum','count'])
sample.columns= ['sales_share', 'game_count_share'] # –ø–µ—Ä–µ–∏–º–µ–Ω—É–µ–º –∫–æ–ª–æ–Ω–∫–∏, —á—Ç–æ–± –Ω–∞ –≥—Ä–∞—Ñ–∏–∫—É –±—ã–ª–æ —è—Å–Ω–æ, –æ —á–µ–º –∏–¥–µ—Ç —Ä–µ—á—å
sample['game_count_share']=sample['game_count_share']/data['total_sales'].count()*100
# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–≥—Ä –≤ –∂–∞–Ω—Ä–µ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫ –≤—ã–≤–µ–¥–µ–º –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö –æ—Ç –≤—Å–µ—Ö –∏–≥—Ä
sample['sales_share']=sample['sales_share']/data['total_sales'].sum()*100
# –í—ã—Ä—É—á–∫—É –æ—Ç –∏–≥—Ä –∫–∞–∂–¥–æ–≥–æ –∂–∞–Ω—Ä–∞ –≤—ã–≤–µ–¥–µ–º –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö –æ—Ç –æ–±—â–µ–π –≤—ã—Ä—É—á–∫–∏
sample.plot(kind='bar', figsize=(15,7),ylabel='–î–æ–ª—è –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö', title='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–≥—Ä –∏ —Å—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –∂–∞–Ω—Ä–∞–º');


data.groupby('genre')['total_sales'].agg(['mean','median']).plot(kind='bar', color=['r', 'g'],title='–°—Ä–µ–¥–Ω–µ–µ –∏ –º–µ–¥–∏–∞–Ω–∞ –ø—Ä–æ–¥–∞–∂ –ø–æ –∂–∞–Ω—Ä–∞–º', figsize=(15,7));
```

–° —É—á–µ—Ç–æ–º –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö, –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –≤—ã–≤–æ–¥—ã:
- –∏–≥—Ä –≤ –∂–∞–Ω—Ä–µ action —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ (–±–æ–ª–µ–µ 35%), –≤ —Å–≤—è–∑–∏ —Å —á–µ–º –∏—Ö –ø—Ä–æ–¥–∞–∂–∏ —Å–æ—Å—Ç–∞–≤–∏–ª–∏ 30%;
- –∏–≥—Ä—ã –≤ –∂–∞–Ω—Ä–µ shooter –ø—Ä–æ–¥–∞—é—Ç—Å—è –Ω–∞–º–Ω–æ–≥–æ –ª—É—á—à–µ –¥—Ä—É–≥–∏—Ö: –¥–æ–ª—è –∏—Ö –æ—Ç —á–∏—Å–ª–∞ –∏–≥—Ä –º–µ–Ω–µ–µ 10%, –æ–¥–Ω–∞–∫–æ –ø–æ—á—Ç–∏ —á–µ—Ç–≤–µ—Ä—Ç—å –≤—Å–µ—Ö –ø—Ä–æ–¥–∞–∂ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –∑–∞ –∏—Ö —Å—á—ë—Ç;
- –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ –ø–æ 15% –ø—Ä–æ–¥–∞–∂ –ø—Ä–∏—Ö–æ–¥–∏—Ç—Å—è –Ω–∞ —Ä–æ–ª–µ–≤—ã–µ –∏ —Å–ø–æ—Ä—Ç–∏–≤–Ω—ã–µ –∏–≥—Ä—ã;
- –ø—Ä–∏–∫–ª—é—á–µ–Ω–∏—è, —Å—Ç—Ä–∞—Ç–µ–≥–∏–∏ –∏ –ø–∞–∑–ª—ã –ø—Ä–æ–¥–∞—é—Ç—Å—è —Ö—É–∂–µ –¥—Ä—É–≥–∏—Ö –∏–≥—Ä;
- –≤–æ –≤—Å–µ—Ö –∂–∞–Ω—Ä–∞—Ö —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–∂ –∑–Ω–∞—á–∏–º–æ –ø—Ä–µ–≤—ã—à–∞–µ—Ç –º–µ–¥–∏–∞–Ω—É –ø—Ä–æ–¥–∞–∂, –∏–∑ —á–µ–≥–æ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥ –æ –Ω–∞–ª–∏—á–∏–∏ –±–µ—Å—Ç—Å–µ–ª–ª–µ—Ä–æ–≤ - –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –∏–≥—Ä, —Å–æ—Å—Ç–∞–≤–ª—è—é—â–∏—Ö "–≤—ã–±—Ä–æ—Å—ã" –≤ –±–æ'–ª—å—à—É—é —Å—Ç–æ—Ä–æ–Ω—É.

## –ê–Ω–∞–ª–∏–∑ –ø–æ–≤–µ–¥–µ–Ω–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º

### –ü–æ–ø—É–ª—è—Ä–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã

```python
sample = data.pivot_table(index='platform', values =['na_sales', 'eu_sales', 'jp_sales', 'other_sales'],aggfunc='sum')
# —Å–æ–∑–¥–∞–Ω–∞ —Ç–∞–±–ª–∏—Ü–∞ –ø—Ä–æ–¥–∞–∂, —Å–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ, –ø–æ –≤—Å–µ–º —Ä–µ–≥–∏–æ–Ω–∞–º.
sum_sample = sample.sum() 
# –ø–æ–ª—É—á–µ–Ω—ã —Å—É–º–º–∞—Ä–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –ø–µ—Ä–µ–≤–æ–¥–∞ –∞–±—Å–æ–ª—é—Ç–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ –¥–æ–ª–∏

for item in list(sample.columns):
    sample[item] = sample[item] / sum_sample[item] * 100 # –ø–µ—Ä–µ–≤–æ–¥–∏–º –∞–±—Å–æ–ª—é—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –ø—Ä–æ—Ü–µ–Ω—Ç—ã
    
```

```python
display(sample)
sample.plot(kind='bar',figsize = (16, 6));
```

–ö–∞–∫ –≤–∏–¥–Ω–æ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ, –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∫–æ–Ω—Å–æ–ª–µ–π —Å–∏–ª—å–Ω–æ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç —Ä–µ–≥–∏–æ–Ω–∞, —á—Ç–æ –º–æ–∂–µ—Ç –±—ã—Ç—å —Å–≤—è–∑–∞–Ω–æ –∫–∞–∫ —Å –æ–±—ä–µ–º–æ–º –ø—Ä–æ–¥–∞–∂ –∫–æ–Ω—Å–æ–ª–∏ –≤ —Ä–∞–∑–Ω—ã—Ö —Å—Ç—Ä–∞–Ω–∞—Ö, —Ç–∞–∫ –∏ —Å –ø—Ä–∏–≤–µ—Ä–∂–µ–Ω–Ω–æ—Å—Ç—å—é –º–µ—Å—Ç–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∫ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–º—É —Ñ–æ—Ä–º–∞—Ç—É –ø–ª–∞—Ç—Ñ–æ—Ä–º. –ù–∏–∂–µ –≤—ã—è—Å–Ω–∏–º –ø–æ–¥—Ä–æ–±–Ω–µ–µ, –∫–∞–∫–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –ø–æ–ø—É–ª—è—Ä–Ω—ã –≤ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö —Ä–µ–≥–∏–æ–Ω–∞—Ö –∏ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –¥–æ–ª–∏ –ø–æ–∫—É–ø–æ–∫ –∏–≥—Ä –¥–ª—è —ç—Ç–∏—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º.  

```python
def get_top_5(df, column):
    series = df[column].sort_values()
    series['other']=100 - series.tail(5).sum()
    series=series.tail(6)
    print('–†–∞—Å—á—ë—Ç top-5 –¥–ª—è', column)
    print(series)
    series.plot(kind='pie', ylabel='–î–æ–ª—è –ø—Ä–æ–¥–∞–∂ –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö', title=column, grid=True, figsize=(10,10));
    plt.show()

regions = sample.columns.to_list()

for region in regions:
    get_top_5(sample, region)
```

–ù–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –º–æ–∂–Ω–æ —É—Ç–≤–µ—Ä–∂–¥–∞—Ç—å, —á—Ç–æ –≤ –Ø–ø–æ–Ω–∏–∏ –Ω–∞–∏–±–æ–ª–µ–µ –ø–æ–ø—É–ª—è—Ä–Ω–∞ –∫–æ–Ω—Å–æ–ª—å 3DS —Å –±–æ–ª—å—à–∏–º –æ—Ç—Ä—ã–≤–æ–º –æ—Ç –∫–æ–Ω–∫—É—Ä–µ–Ω—Ç–æ–≤, –≤ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ - X360, –∑–∞ –∫–æ—Ç–æ—Ä—ã–º —Å–ª–µ–¥—É—é—Ç PS3, PS4, XOne —Å –Ω–µ–±–æ–ª—å—à–∏–º –æ—Ç—Ä—ã–≤–æ–º, –≤ –¥—Ä—É–≥–∏—Ö —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º—ã—Ö —Ä–µ–≥–∏–æ–Ω–∞—Ö –ª–∏–¥–∏—Ä—É—é—Ç PS4, –∏ –∑–∞ –Ω–∏–º —Å –Ω–µ–±–æ–ª—å—à–æ–π —Ä–∞–∑–Ω–∏—Ü–µ–π PS3.

### –ü–æ–ø—É–ª—è—Ä–Ω—ã–µ –∂–∞–Ω—Ä—ã
–†–∞—Å—Å—á–∏—Ç–∞–µ–º –≤–æ—Å—Ç—Ä–µ–±–æ–≤–∞–Ω–Ω—ã–µ –∂–∞–Ω—Ä—ã –∏–≥—Ä –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º –∞–Ω–∞–ª–æ–≥–∏—á–Ω–æ –ø–æ–∏—Å–∫—É –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –∫–æ–Ω—Å–æ–ª–µ–π:

```python
sample = data.pivot_table(index='genre', values =['na_sales', 'eu_sales', 'jp_sales', 'other_sales'],aggfunc='sum')
# —Å–æ–∑–¥–∞–Ω–∞ —Ç–∞–±–ª–∏—Ü–∞ –ø—Ä–æ–¥–∞–∂, —Å–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø–æ –∂–∞–Ω—Ä—É, –ø–æ –≤—Å–µ–º —Ä–µ–≥–∏–æ–Ω–∞–º.
sum_sample = sample.sum() 
# –ø–æ–ª—É—á–µ–Ω—ã —Å—É–º–º–∞—Ä–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –ø–µ—Ä–µ–≤–æ–¥–∞ –∞–±—Å–æ–ª—é—Ç–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ –¥–æ–ª–∏

for item in list(sample.columns):
    sample[item] = sample[item] / sum_sample[item] * 100 # –ø–µ—Ä–µ–≤–æ–¥–∏–º –∞–±—Å–æ–ª—é—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –ø—Ä–æ—Ü–µ–Ω—Ç—ã
    
display(sample)
sample.plot(kind='bar',figsize = (16, 6));
    
```

–£–∂–µ –≤–∏–¥–Ω–æ, —á—Ç–æ –≤–æ –≤—Å–µ–º –º–∏—Ä–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –æ—Ö–æ—Ç–Ω–æ –ø–æ–∫—É–ø–∞—é—Ç action-–∏–≥—Ä—ã, –≤ –Ø–ø–æ–Ω–∏–∏ - RPG, –∞ –≤ –ø—Ä–æ—á–∏—Ö —Ä–µ–≥–∏–æ–Ω–∞—Ö - —à—É—Ç–µ—Ä—ã –∏ —Å–ø–æ—Ä—Ç–∏–≤–Ω—ã–µ –∏–≥—Ä—ã. –í–Ω–æ–≤—å —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º —Ä–µ–≥–∏–æ–Ω—ã –ø–æ –æ—Ç–¥–µ–ª—å–Ω–æ—Å—Ç–∏.

```python
for region in regions:
    get_top_5(sample, region)
```

```python
# –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ 2
games_fresh = data.copy()
countries = games_fresh.columns[4:7]
# –Ω–∞—á–µ—Ä—Ç–∏–º –∫—Ä—É–≥–æ–≤—É—é –¥–∏–∞–≥—Ä–∞–º–º—É –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–µ–≥–∏–æ–Ω–∞
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('–¢–û–ü-5 –∂–∞–Ω—Ä–æ–≤ –≤ —Ä–µ–≥–∏–æ–Ω–∞—Ö')
for i in range(3):
  # –≤—ã–±–µ—Ä–µ–º 5 —Å–∞–º—ã—Ö –ø—Ä–æ–¥–∞—é—â–∏—Ö –∂–∞–Ω—Ä–æ–≤
  df_i = (games_fresh.groupby('genre')[countries[i]].sum().
             reset_index().
             sort_values(countries[i], ascending = False, ignore_index = True).
             head(5))
  # –∑–∞–Ω–µ—Å–µ–º –≤—Å–µ –æ—Å—Ç–∞–ª—å—ã–Ω–µ –ø—Ä–æ–¥–∞–∂–∏ –≤ –¥—Ä—É–≥–∏–µ –∂–∞–Ω—Ä—ã
  df_i = df_i.append({'genre': 'Other_genre', countries[i]: games_fresh[countries[i]].sum() - df_i[countries[i]].sum()}, ignore_index= True)
  axs[i].pie(df_i[countries[i]], labels = df_i['genre'], autopct='%1.1f%%')
  # –ó–∞–¥–∞–¥–∏–º –Ω–∞–∑–≤–∞–Ω–∏–µ —Ä–µ–≥–∏–æ–Ω–∞.
  axs[i].set_title(df_i.columns[1].split('_')[0].upper())
plt.show()
```

```python
sample = data.pivot_table(
    index='rating', values =['na_sales', 'eu_sales', 'jp_sales', 'other_sales'],aggfunc='mean')

sample = sample.rename({'M':'M2'}).sort_index()
sample.loc['M', :]=sample.loc['M2', : ]

sample = sample.drop('M2') # –ø–µ—Ä–µ—Å—Ç–∞–≤–∏–º –≤ —Ç–∞–±–ª–∏—Ü–µ —Å—Ç—Ä–æ–∫–∏, —á—Ç–æ–± –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–Ω–æ–≥–æ —Ä–µ–π—Ç–∏–Ω–≥–∞

for column in sample.columns:
    sample[column].plot(kind='', title = column, figsize = (8,5))
    plt.show()
```
